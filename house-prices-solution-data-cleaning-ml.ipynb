###### display the columns with the number of missing values

```python
data.isnull().sum().sort_values(ascending=False)[:34]
```

 we will impute and clean the missing values:
1. impute the mean value to certain numerical features.
2. create a new class 'other' for certain categorical features.
3. use the mode value for the other categotical features.
4. drop the ID and certain unnecessary features

```python
data = data.drop(columns=['Id','Street','PoolQC','Utilities'],axis=1)
```

```python
data['LotFrontage'] = data.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))
```

```python
data['LotFrontage'].isnull().sum()
```

```python
#create a new class 'other'
features=['Electrical','Functional','KitchenQual','SaleType','Exterior2nd','Exterior1st','MiscFeature','Alley','Fence','FireplaceQu','GarageCond','GarageQual','GarageFinish','GarageType','BsmtCond','BsmtExposure','BsmtQual','BsmtFinType2','BsmtFinType1','MasVnrType']
for name in features:
    data[name].fillna('Other',inplace=True)
```

```python
data[features].isnull().sum()
```

```python
data['MSZoning'] = data.groupby('MSSubClass')['MSZoning'].transform(lambda x: x.fillna(x.mode()[0]))
```

```python
zero=['GarageYrBlt','GarageArea','MasVnrArea','BsmtHalfBath','BsmtHalfBath','BsmtFullBath','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','GarageCars']
for name in zero:
    data[name].fillna(0,inplace=True)
```

```python
data.isnull().sum().sum()
```

dataset no longer has any missing values. 

 transform categorical features into numerical features

```python
data.select_dtypes(include='object').columns
```

```python
def dummies(d):
    dummies_df=pd.DataFrame()
    for name in d.select_dtypes(include='object').columns:
        dummies = pd.get_dummies(d[name], drop_first=False)
        dummies = dummies.add_prefix("{}_".format(name))
        dummies_df=pd.concat([dummies_df,dummies],axis=1)
    return dummies_df
```

```python
dummies_data=dummies(data)
dummies_data.shape
```

```python
object_features=['MSZoning','Alley', 'LotShape', 'LandContour',
       'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
       'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
       'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation',
       'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
       'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
       'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual',
       'GarageCond', 'PavedDrive','Fence', 'MiscFeature',
       'SaleType', 'SaleCondition']
```

```python
data_drop=data.drop(columns=object_features,axis=1)
data_drop.shape
```

```python
final_data=pd.concat([data_drop,dummies_data],axis=1)
final_data.shape
```

##### Regression algorithms 

```python
#Re-spliting the data into train and test datasets
train_data=final_data.iloc[:1460,:]
test_data=final_data.iloc[1460:,:]
print(train_data.shape)
test_data.shape
```

```python
# X: independent variables and y: target variable
X=train_data
y=train.loc[:,'SalePrice']
```

```python
from sklearn.linear_model import Ridge, RidgeCV, LassoCV
```

```python
import xgboost as xgb
```

```python
model_xgb = xgb.XGBRegressor(learning_rate=0.01,n_estimators=3460,
                                     max_depth=3, min_child_weight=0,
                                     gamma=0, subsample=0.7,
                                     colsample_bytree=0.7,
                                     objective='reg:linear', nthread=-1,
                                     scale_pos_weight=1, seed=27,
                                     reg_alpha=0.00006)
model_xgb.fit(X, y)
xgb_preds=model_xgb.predict(test_data)
```

```python
xgb_preds
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
model_gbr = GradientBoostingRegressor(n_estimators=3000, 
                                learning_rate=0.05, 
                                max_depth=4, 
                                max_features='sqrt', 
                                min_samples_leaf=15, 
                                min_samples_split=10, 
                                loss='huber', 
                                random_state =42)
model_gbr.fit(X, y)
gbr_preds=model_gbr.predict(test_data)
```

```python
predictions = 0.35 * xgb_preds + 0.65 * gbr_preds
```

```python
#display the first 5 predictions of sale price
predictions[:5]
```

```python
#make the submission data frame
submission = {
    'Id': test.Id.values,
    'SalePrice': predictions
}
solution = pd.DataFrame(submission)
solution.head()
```

```python
#make the submission file
solution.to_csv('submission.csv',index=False)
solution
```
